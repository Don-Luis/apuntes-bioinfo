<h3>plot</h3>
(+info: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.plot"  target="_blank">plot function</a>)

<p>The plot function is generally used as:</p>

<pre>plot(Y [, ’line options’]) 
plot(X1, Y1 [, ’line 1 options’], X2, Y2[, ’line 2 options ’], ...)
plot(..., property=Value,...)</pre>

<p>This function plots the values of the array Yn with respect to Xn. Optionally, different line styles can be used for each pair of Xn and Yn arrays. If no line style is given then matlablib uses a different style for each line. Finally, you can assign properties to the lines of the plot using pairs of <tt>property=value</tt> in the arguments.</p>

<p>Let's see some options without line options:</p>

```python
# Simple plot of a vector with respect to anothe vector
x = np.linspace(0.1, 4, 40)
y = 1/x
plt.plot(x, y)
```

```python
# Plot of sin and cosine function with respect to the same x

x = np.linspace(-np.pi, np.pi, 40) 
y1 = np.sin(x)
y2 = np.cos(x)
# Use dummy variable _ to get the return value of plot to avoid printing its reference
_ = plt.plot(x, y1, x, y2)
```

<h4>Line options</h4>(+info in: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.plot"  target="_blank">plot function</a>)

You can specify line option for the lines of the plot. For that specify a string after each pair of Xn and Yn parameters.



```python
# Simple plot with purple stars

x = np.arange(0.1,4,0.1)
y = 1/x
plt.plot(x, y, 'm*')
```

```python
# Different types of line

x  = np.arange(0.1,20,0.1)
y  = np.exp(0.1*x)
y1 = y * np.sin(x)
y2 = y * np.cos(x)
_ = plt.plot (x, y, 'k', x, -y, 'k', x, y1, 'k--', x, y2, 'k:')
```

```python
# Different markers

x  = np.arange(0.1,20,0.5)
y  = np.exp(0.1*x)
y1 = y * np.sin(x)
y2 = y * np.cos(x)
plt.plot (x, y, 'kv', x, -y, 'k^', x, y1, 'k*', x, y2, '.k')
```

<h4>Line properties</h4> (+info: <a href="http://matplotlib.org/1.5.0/api/lines_api.html#matplotlib.lines.Line2D" target="_blank">2D line</a>)
<p>Other line properties can be assigned using property-value pairs. This properties are assigned to all lines in the plot. Some of the most useful are:</p>
<ul>
<li><tt>linewidth</tt>: Line width in pixels.</li>
<li><tt>markeredgecolor</tt>: Marker edge color.</li>
<li><tt>markerfacecolor</tt>: Marker face color.</li>
<li><tt>markersize</tt>: Size of the marker.</li>
</ul>

```python
#Other properties

x  = np.arange(0.1,20,0.5)
y  = np.exp(0.1*x)
y1 = y * np.sin(x)
y2 = y * np.cos(x)
# The specified properties affect all lines of the plot but...see next example
_ = plt.plot (x, y1, ':ms', x, y2, '-co', linewidth=2, markeredgecolor='k', markerfacecolor='r', markersize=7)
```

```python
# To set different properties to different lines you have to call plot several times

x  = np.arange(0.1,20,0.5)
y  = np.exp(0.1*x)
y1 = y * np.sin(x)
y2 = y * np.cos(x)

_ = plt.plot (x, y1, ':ms', linewidth=2, markeredgecolor='k', 
              markerfacecolor='r', markersize=5)
_ = plt.plot (x, y2, '-co', linewidth=11, markeredgecolor='b', 
              markerfacecolor='c', markersize=11)
```

<h3>Pie</h3>

<h3>hist</h3>
<p>The forms of this function are:</p>
<pre>
hist(X) 
hist(X, nbins)
</pre>
<p>This function compute and plot the histogram of the values in X. If nbins is not specified hist creates 10 bins equally spaced. If X is a matrix then one histogram per column is ccomputed and ploted.</p>

```python
# Histogram

# Generate 10000 random numbers with normal distribution
x = np.random.randn(10000)
_ = plt.hist(x)
```

```python
#Histogram with 50 bins

x = np.random.randn(10000)
_ = plt.hist(x, 50)
```

```python
# Histogram of a matrix

# Generate 10000 random numbers with normal distribution and 10000 with 
# uniform distribution from -3 to 3
X = np.array([np.random.randn(10000), np.random.rand(10000)*6-3])
_ = plt.hist(X.T, 25)
```

<h3>bar and barh</h3>
(+info: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.bar" target="_blank">function bar</a>)
<p>These functions are used as:</p>
<pre>
bar(x,y, width=0.8, bottom=None, style...)
</pre>
<p>These function plot a graph with vertical bars (bar) or horizontal (barh):
The parameter ‘width’ indicates the separation bertween bars. The bottom parameter can be used to create stacked bar plots.</p>

```python
# Barras verticales

x = np.arange(1., 10., 1)
y = 1/x
_ = plt.bar(x,y)
```

```python
# Horizontal bars

x = np.arange(1., 10., 1)
y = 1/x
_ = plt.barh(x, y)
```

```python
# Other options

# Stacking

x = np.arange(1., 10., 1)
y = 1/x
plt.bar(x, y, width = 1, color='r')
plt.bar(x, y[::-1], width = 1, bottom=y)

```

<h3>contour</h3>
(+info: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.contour" target="_blank">contour function</a>)
<p>Forms:</p>
<pre>
contour(Z) 
contour(Z, n)
</pre>
<p>This function plots isolines from matrix Z, where Z is interpreted as heights with respect to the xy plane. Z should be at least a 2x2 matrix. The number of levels can be specified using 'n' parameter, otherwise they are calculated automatically based on the max and min heights in Z. </p>

```python
#Isolines

x = np.arange(-20,20,0.5)
Z = np.array(np.matrix(x).transpose()*np.matrix(x))
_ = plt.contour(Z)
```

```python
#Isolines with 25 levels

x = np.arange(-20,20,0.5)
Z = np.array(np.matrix(x).transpose()*np.matrix(x))
plt.contour(Z, 25)
```

<h3>pcolor</h3>
(+info: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.pcolor" target="_blank">function pcolor</a>)
<pre>
pcolor(C) 
</pre>
<p>It draws a colored matrix with colors given by C. The plot shows a column and row less than the columns and rows in the matrix. This functions (if no other parameters are given) gets the color for cell (i, j) from the element (i, j) of the matrix C using the actual colormap (+info: help(plt.cm)).</p>

```python
#Celdas coloreadas
x = np.arange(-5,5,0.5)
C = np.array(np.matrix(x).transpose()*np.matrix(x))
plt.pcolor(C)
```

```python
#Green scaled cells
x = np.arange(-5,5,0.5)
C = np.array(np.matrix(x).transpose()*np.matrix(x))
plt.pcolor(C, cmap=plt.cm.Greens)
```

<h3>scatter</h3>
(+info: <a href="http://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.scatter" target="_blank">scatter function</a>)
<p>Form:</p>
<pre>
scatter(x, y, s=20, c=u'b',...) 
</pre>
<p>Scatter plot with coordinates x and y. Vectors x and y have to be of the same size. The size of the dots can be specified using <tt>s</tt> and their color with <tt>c</tt>. Both size and color can be given for all points together using a single value or for each single point using a vector. For the first configuration, it is enough to give a single scalar for the size and a string for the color. For the second one, a vector of the same size of x and y should be given for the sizes. For the colors, you can specify a vector containing indixes f the colormap, or an array of size len(x) x 3 with colors in RGB.</p>

```python
# Scatter plot of random normally distributed points in 2D 
n_points = 100
x = np.random.randn(n_points)
y = np.random.randn(n_points)
_ = plt.scatter(x, y)
```

```python
#...with bigger points

_ = plt.scatter(x, y, s=100)
```

```python
#...with orange dots of random sizes

_ = plt.scatter(x, y, np.random.rand(n_points)*500, c='orange')
```

```python
#...with random colors using the colormap

_ = plt.scatter(x, y, s=100, c=np.random.rand(n_points))
```

<h3>Global options for plots</h3>
<p>Some important functions to create a "proper" plot are:</p>
<ul>
<li>figure: Creates a figure. Usually it is not necessary since the figure is created automatically when calling any of the previous plotting functions. However, it could be interesting to call it to change some of the default graph options like size, background color, etc.</li>
<li>subplot(nrows, ncols, index): To create a matrix of graphs of nrows x ncols. The last parameter indicates over which plot we will work. </li>
<li>xlabel and ylabel: To set the labels for the X and Y axis.</li>
<li>title: To set the title of the plot.</li>
<li>legend: To add a legend to the plot. It has to be called after executing the plotting function.</li>
<li>colorbar: To include the color bar for reference.</li>
<li>xlim and ylim: To modify the limits to show in the plot.</li>
<li>xticks y yticks: To modify the labels of each value of the axis.</li>
<li>...</li>
</ul>
<p>The next example shows how to use these functions:</p>

```python
plt.figure(num=None, figsize=(12, 4), facecolor='c')

# Data
x = np.linspace(-np.pi,np.pi,20)
Y1 = np.sin(x)
Y2 = np.cos(x)

# Graph using legend and labels in the axis
plt.subplot(1,2,1) # Create a graph with two subplots. We start working on the first
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, Y1, label='sen')
plt.plot(x, Y2, label='cos')
plt.legend(loc=2) # Called after finishing the plot

# Graph with x label, title, with the limits of the axis modified
# and also the labels for the x axis
plt.subplot(1,2,2) # Now we use the second subplot
plt.xlabel('x')
plt.title('Seno y coseno')
plt.xlim([-3,3])
plt.xticks(range(-3,4), ['a','b','c','d','e','f','g'])
plt.plot(x, Y1, label='sen')
_ = plt.plot(x, Y2, label='cos')
```

```python
# Color reference bar
plt.figure(figsize=(4.7, 4))
x = np.arange(-5,5,0.5)
C = np.array(np.matrix(x).transpose()*np.matrix(x))
plt.pcolor(C)
plt.colorbar() # Has to be called after ploting
```

```python
def sigmoid(V, V0, W=12):
    """ Sigmoidal function
    
    Parameters
    ----------
    V: float or array of floats
        Membrane potential
    V0: float
        Position of the sigmoid function
    W: float
        Width of the transition region
        
    Returns
    -------
    y: float or array of floats
        Sigmoidal function    
    """
    return 1/(1+np.exp(-(V-V0)/W))
```

```python
V = np.arange(-100, 100, 1) #if we use linspace, instead of the steps we put how many values we want
y = sigmoid(V, -40, 12)

fig, axs = plt.subplots(1, 2)
#varying V0
ax = axs[0]
for V0 in [-40, 0, 40]:
    y = sigmoid(V, V0, 12)
    ax.plot(V, y, label = f'{V0:.0f}mV')
ax.set_xlabel('V [mV]')
ax.legend(title = '$V_0$')


#Varying W 
ax = axs[1]
for w in [1, 10, 20]:
    y = sigmoid(V, 0, w)
    ax.plot(V, y, label = f'{w:.0f}mV')
ax.set_xlabel('V [mV]')
ax.legend(title = '$w$')
plt.show()
```
